```python
from pandas import *
import cufflinks as cf; cf.go_offline()
df=cf.datagen.lines(4,mode='abc')
df.head()
```

## Line plots

```python
df.iplot(y='a')
```

```python
df.iplot(secondary_y=['c', 'd'])
```

```python
df.iplot(subplots=True, subplot_titles=['test1','test2','test3','test4'], legend=False, title = 'Check it out')
```

# Bars

```python
df.iplot(kind='bar')
```

```python
df.iplot(kind='bar', barmode='stack')
```

## Histograms

```python
df.iplot(kind='histogram', subplots=True, shape=(4, 1), filename='histogram-subplots')
```
